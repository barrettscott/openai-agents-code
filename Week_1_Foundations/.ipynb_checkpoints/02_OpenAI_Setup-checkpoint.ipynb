## 🔑 Step 1: Create OpenAI Account & Add Credits

1. Go to [OpenAI Platform](https://platform.openai.com/signup)
2. Create an account (or sign in if you have one)
3. Verify your email

### Add Payment & Credits

OpenAI API usage requires prepaid credits.

1. Go to [OpenAI Billing](https://platform.openai.com/account/billing/overview)
2. Click **"Add payment method"** and enter your credit card
3. Click **"Add to credit balance"** and add at least `$5` 

💡 **Set budget alerts (recommended):**

- Go to Settings → Limits → [Usage limits](https://platform.openai.com/settings/organization/limits)
- Set a monthly budget (e.g., `$10`)
- ⚠️ **Note:** This is a soft limit for alerts only — it won't automatically stop requests

---

### Get Your API Key

1. Go to [OpenAI API Keys](https://platform.openai.com/api-keys)
2. Click **"Create new secret key"**
3. Name it "openai-agents"
4. Complete any verification prompts if shown
5. **Copy the key immediately** — you can't see it again!

**Example key format:**
sk-proj-abcdefghijklmnopqrstuvwxyz1234567890...

⚠️ **Security:** Never share your API key with anyone or post it online. Treat it like a password!

## 📁 Step 2: Create Your .env File

### What is a .env file?

A `.env` file stores your API key securely:
- Keeps secrets out of your code
- Standard practice in professional development
- Never gets committed to version control

**Where it goes:**
```
📁 openai-agents/
├── 📁 M01_Setup/
│   └── 📓 OpenAI Setup_OpenAI_Setup.ipynb  ← You are here
├── 🔐 .env               ← We'll create this here
└── 📄 environment.yml
```

💡 **Note:** Files starting with `.` are hidden by default on macOS and Linux.
- **macOS:** In Finder, press **Cmd + Shift + .** to toggle hidden files
- **Linux:** In your file manager, press **Ctrl + H** to toggle hidden files
- **Windows:** Hidden files starting with `.` are visible by default

### Create your .env file

Run the cell below. When prompted, paste your API key and press Enter.

*Already have a `.env` file? The cell will detect it and ask if you want to keep or replace it.*

*Input box not appearing? Use the Manual Alternative below.*

````python
from pathlib import Path
from getpass import getpass

# Path to .env file (in parent directory)
env_file = Path("..") / ".env"

print("Creating .env file...\n")
print(f"Location: {env_file.resolve()}\n")

# Check if .env already exists
if env_file.exists():
    print("ℹ️  .env file already exists!")
    print(f"   Location: {env_file.resolve()}\n")
    
    response = input("Do you want to replace it? (yes/no): ").strip().lower()
    
    if response not in ['yes', 'y']:
        print("\n✅ Keeping existing .env file.")
        print("   If you need to update it later, delete it manually and run this cell again.")
    else:
        env_file.unlink()  # Delete the file
        print("\n✅ Old .env file deleted. Creating new one...\n")

# If we get here, either file didn't exist OR user chose to replace it
if not env_file.exists():
    api_key = getpass("Paste your API key (hidden): ").strip()
    
    # Validation
    if not api_key:
        print("\n❌ No API key entered.")
        print("   Run this cell again and paste your key.")
    elif not api_key.startswith("sk-"):
        print("\n⚠️ Warning: API key doesn't start with 'sk-'")
        print("   Make sure you copied the entire key!")
    else:
        # Create .env file
        with open(env_file, 'w') as f:
            f.write(f"OPENAI_API_KEY={api_key}\n")
        
        print("\n✅ .env file created successfully!")
        print(f"   Saved to: {env_file.resolve()}")
        print("\n🔒 Your API key is stored securely.")
````

## 🔍 Step 3: Verify Your Setup

Let's check that your .env file was created correctly:

````python
import os
from pathlib import Path
from dotenv import load_dotenv

print("Verifying .env file...\n")

# Load .env file
env_path = Path("..") / ".env"
load_dotenv(dotenv_path=env_path, override=True)

# Check if API key exists
api_key = os.getenv("OPENAI_API_KEY")

if not api_key:
    print("❌ API key not found!\n")
    print("Troubleshooting:")
    print("1. Make sure you ran the previous cell OR created .env via terminal")
    print("2. Check that .env file exists in the openai-agents folder")
    print("3. Check .env contains: OPENAI_API_KEY=your-key")
    print("4. No spaces around the = sign")
    print("5. Windows users: recreate .env using the PowerShell command in Step 2")
elif "your-actual-key-here" in api_key or api_key == "sk-proj-your-actual-key-here":
    print("❌ Placeholder key detected!\n")
    print("You need to replace 'sk-proj-your-actual-key-here' with your real API key.")
    print("Go back to Step 2 and paste your actual key from OpenAI.")
elif not api_key.startswith("sk-"):
    print("⚠️ API key doesn't start with 'sk-'\n")
    print("Make sure you copied the entire key from OpenAI")
else:
    # Show masked key
    masked_key = f"{api_key[:8]}...{api_key[-4:]}"
    print(f"✅ API key loaded: {masked_key}")
    print(f"✅ Key length: {len(api_key)} characters")
    print("\n🔒 Your full key is hidden for security")


    print("\n✅ Ready to test the connection!")
````

## 🚀 Step 4: Verify Your Connection with Your First Agent Call

Let's test everything by asking the AI a question:

````python
from pathlib import Path
from dotenv import load_dotenv
load_dotenv(dotenv_path=Path("..") / ".env", override=True)

from agents import Agent, Runner

MODEL = "gpt-5-mini"

agent = Agent(
    name="SetupTest",
    instructions="You are a helpful assistant.",
    model=MODEL
)

try:

    result = await Runner.run(agent, input="Say 'Setup complete!' and nothing else.")

    print("🎉 SUCCESS! Your first agent response:\n")
    print(result.final_output)
    print("\n✅ You're ready for the course!")

except Exception as e:
    print(f"❌ Error: {e}\n")
    error_msg = str(e).lower()
    if "authentication" in error_msg or "api key" in error_msg:
        print("→ Check your API key is correct in .env")
    elif "billing" in error_msg or "quota" in error_msg:
        print("→ Add credits: https://platform.openai.com/account/billing/overview")
    elif "rate limit" in error_msg:
        print("→ Wait 60 seconds and try again")
    else:
        print("→ See troubleshooting section below")
````

## ✅ Setup Complete: Final Check

You've made a successful agent call! Let's confirm everything is ready for the rest of the course:

````python
import os
from pathlib import Path
from dotenv import load_dotenv

print("=" * 60)
print("🎓 FINAL SETUP CHECK")
print("=" * 60)

checks = []

# .env location
env_path = Path("..") / ".env"
env_ok = env_path.exists()
checks.append((".env file exists", env_ok))

# Load environment
if env_ok:
    load_dotenv(dotenv_path=env_path, override=True)

# openai-agents package
try:
    from agents import Agent, Runner
    checks.append(("openai-agents installed", True))
except ImportError:
    checks.append(("openai-agents installed", False))

# API Key (presence + basic format only)
api_key = os.getenv("OPENAI_API_KEY", "")
key_ok = bool(api_key) and api_key.startswith("sk-")
checks.append(("API key present (looks correct)", key_ok))

# Print results
print()
for name, passed in checks:
    status = "✅" if passed else "❌"
    print(f"{status} {name}")

print()
if all(passed for _, passed in checks):
    print("=" * 60)
    print("🎉 YOU'RE READY FOR THE COURSE!")
    print("=" * 60)
    print("\nYour environment is configured correctly.")
else:
    print("❌ Some checks failed. Fix the ❌ items above.")
````

## 🔧 Troubleshooting

**Authentication error or API key not found?**
- Check you copied the entire key (starts with `sk-`)
- Format: `OPENAI_API_KEY=sk-your-key` (no spaces around `=`)
- Check `.env` is in the `openai-agents` folder, not inside `M01_Setup`

**Key not loading after editing `.env`?**
- Restart the kernel after editing `.env` — changes don't take effect until the kernel restarts
- Windows users: recreate `.env` using the PowerShell command in the Manual Alternative section

**Billing error (no credits)?**
- Add payment method at platform.openai.com/account/billing
- Minimum $5 credit required

**.env file not found?**
- File must be named exactly `.env` (starts with dot)
- Must be in the `openai-agents` folder, not inside `M01_Setup`

**Connection or rate limit error?**
- Check internet connection and disable VPN temporarily
- Wait 60 seconds if you hit rate limits
- Check status.openai.com for outages

**Still having issues?**
- Copy any error message and paste it into ChatGPT, Claude, Gemini, or Grok — they're great at debugging setup issues
- Re-watch the setup lecture
- Post to the Q&A with your error message and output
