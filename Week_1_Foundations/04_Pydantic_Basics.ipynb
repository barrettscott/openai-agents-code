# Pydantic Basics

*Notebook 04*

Pydantic defines and validates a schema for Python data.

In agent workflows, it catches schema-invalid output before downstream code uses it.

---

## 🔧 Setup

```python
from typing import Annotated
from pydantic import BaseModel, Field, ValidationError

print("✅ Ready!")
```

---

## 📋 Part 1: Your First Pydantic Model

A Pydantic model inherits from `BaseModel`.

Type hints define its fields.

```python
class Person(BaseModel):
    name: str
    age: int
    email: str

# --------------------------------------------------------------
print("✅ Person model ready")
```

Creating an instance looks like normal class construction.

Pydantic validates fields and may coerce compatible values.

```python
person = Person(name="Alice", age="30", email="alice@example.com")

print(person)
print(f"age type: {type(person.age).__name__}")
```

---

## ⚠️ Part 2: Validation in Action

If Pydantic cannot validate a required field, it raises a `ValidationError`.

The error identifies the field and the failed rule.

In agent workflows, this catches schema-invalid output before downstream code uses it.

#### Invalid: Value Cannot Be Parsed

```python
try:
    person = Person(name="Bob", age="thirty", email="bob@example.com")
except ValidationError as e:
    print(e)
```

#### Invalid: Missing Field

```python
try:
    person = Person(name="Carol", age=25)
except ValidationError as e:
    print(e)
```

---

## 📏 Part 3: Field Constraints

Type hints declare the basic type: string, integer, boolean.

For tighter rules, wrap the type in `Annotated[type, Field(...)]`.

Examples: a rating from 1 to 5, or a name under 50 characters.

`Annotated` adds constraints without changing the underlying type.

`ge` and `le` mean greater-or-equal and less-or-equal.

```python
class ProductReview(BaseModel):
    product_name: Annotated[str, Field(max_length=50)]
    rating: Annotated[int, Field(ge=1, le=5)]
    comment: Annotated[str, Field(max_length=200)]

# --------------------------------------------------------------
print("✅ ProductReview model ready")
```

#### Valid Review

```python
review = ProductReview(
    product_name="Wireless Headphones",
    rating=4,
    comment="Great sound quality, comfortable fit."
)

print(review)
```

#### Invalid: Rating Out of Range

```python
try:
    review = ProductReview(
        product_name="Wireless Headphones",
        rating=7,
        comment="Amazing!"
    )
except ValidationError as e:
    print(e)
```

### 💡 Key Takeaway

Validation rules belong in the Pydantic model, not only in prompt wording.

---

## 🔍 Part 4: Accessing and Serializing Models

Once a model instance is created, access fields with standard dot notation.

```python
review = ProductReview(
    product_name="Laptop Stand",
    rating=5,
    comment="Sturdy and adjustable."
)

print(f"Product: {review.product_name}")
print(f"Rating: {review.rating}/5")
print(f"Comment: {review.comment}")
```

Pydantic models serialize to dictionaries and JSON.

Use `model_dump()` when passing data to Python code.

Use `model_dump_json()` when you need a JSON string for an API or file.

```python
print("As dict:")
print(review.model_dump())

print("\nAs JSON:")
print(review.model_dump_json(indent=2))
```

## 🧠 Why This Course Uses Pydantic

The Agents SDK can derive structured-output schemas from Pydantic models and other Python types.

This course uses Pydantic for three reasons:

- **Early failure**: schema-invalid output fails before downstream code uses it

- **Readable models**: Python types are easier to maintain than hand-written schema dictionaries

- **IDE support**: fields such as `review.rating` autocomplete naturally

## 🎯 Key Takeaways

**Pydantic validates data at construction:**

- Inherit from `BaseModel` and define fields with type hints

- Compatible values may be coerced by default

- Invalid or missing fields raise `ValidationError`

- Validation checks schema rules, not factual accuracy
<br>
<br>

**Field constraints add domain rules:**

- Use `Annotated[type, Field(...)]` for bounds and length limits

- Common constraints include `ge`, `le`, `min_length`, and `max_length`

- Failed constraints raise `ValidationError`
<br>
<br>

**Pydantic works naturally with Python:**

- Models provide readable types and IDE autocomplete

- Use `model_dump()` for dictionaries

- Use `model_dump_json()` for JSON strings
