## 📦 Check 4: Required Packages

Let's verify all course packages are installed.

These packages are used in later notebooks — you don't need to know them yet.

```python
import warnings
warnings.filterwarnings("ignore", message="IProgress not found")

print("Checking required packages...\n")

missing = []       # packages not installed at all
wrong_version = [] # packages installed but wrong version

try:
    import agents
    print(f"✅ openai-agents {agents.__version__}")
except ImportError:
    print("❌ openai-agents not found")
    missing.append("openai-agents")

try:
    import openai  # base library, also used by the Agents SDK
    print(f"✅ openai {openai.__version__}")
except ImportError:
    print("❌ openai not found")
    missing.append("openai")

try:
    from dotenv import load_dotenv
    print("✅ python-dotenv")
except ImportError:
    print("❌ python-dotenv not found")
    missing.append("python-dotenv")

try:
    from pydantic import BaseModel
    print("✅ pydantic")
except ImportError:
    print("❌ pydantic not found")
    missing.append("pydantic")

try:
    import chromadb
    print(f"✅ chromadb {chromadb.__version__}")
except ImportError:
    print("❌ chromadb not found")
    missing.append("chromadb")

try:
    import gradio as gr
    major = int(gr.__version__.split(".")[0])
    if major >= 6:  # Week 5 features require Gradio 6+
        print(f"✅ gradio {gr.__version__}")
    else:
        print(f"⚠️  gradio {gr.__version__} — needs 6.x or higher")
        wrong_version.append("gradio")
except ImportError:
    print("❌ gradio not found")
    missing.append("gradio")

try:
    import fastapi
    print(f"✅ fastapi {fastapi.__version__}")
except ImportError:
    print("❌ fastapi not found")
    missing.append("fastapi")

try:
    import uvicorn
    print(f"✅ uvicorn {uvicorn.__version__}")
except ImportError:
    print("❌ uvicorn not found")
    missing.append("uvicorn")

# ---------------------------------------------------------------------------
# Fix instructions
# ---------------------------------------------------------------------------

if missing or wrong_version:
    print()
    print("── Fix ────────────────────────────────────────────────")
    print("Open the VS Code integrated terminal (Ctrl+` / Cmd+`).")
    print()
    if missing:
        print("Missing packages — run:")
        print("    conda env update -n openai-agents -f ../environment.yml")
        print()
    if wrong_version:
        print("Wrong gradio version — run:")
        print("    pip install 'gradio>=6.0'")
        print()
    print("Then: click the kernel indicator (top-right) → Restart, and re-run this cell.")
    print("────────────────────────────────────────────────────────")
else:
    print("\n✅ All packages look good!")
```

## 📌 Check 5: Pin Your Kernel

This writes your interpreter path to `.vscode/settings.json` so VS Code auto-selects `openai-agents` after every reboot. Without pinning, VS Code can silently fall back to a different Python between sessions, which causes confusing 'module not found' errors even though the package is installed.

Run this cell once — VS Code should auto-select `openai-agents` from now on.

```python
import sys
import json
from pathlib import Path

print("Pinning VS Code kernel...\n")

# Guard: only run from inside the openai-agents environment
if "openai-agents" not in sys.executable.lower() and "openai-agents" not in sys.prefix.lower():
    print("⚠️  Not running in the openai-agents environment.")
    print("    Select the openai-agents kernel first, then re-run this cell.")
else:
    settings_path = Path("..") / ".vscode" / "settings.json"
    settings_path.parent.mkdir(exist_ok=True)

    settings = {}
    if settings_path.exists():
        settings = json.loads(settings_path.read_text())

    if settings.get("python.defaultInterpreterPath") == sys.executable:
        print(f"✅ Already pinned to: {sys.executable}")
    else:
        settings["python.defaultInterpreterPath"] = sys.executable
        settings_path.write_text(json.dumps(settings, indent=4))
        print(f"✅ Kernel pinned to: {sys.executable}")
        print("\n    VS Code will now auto-select openai-agents after every reboot.")
        print("    No manual kernel selection needed going forward.")
```
