# How Agents Work

*Notebook 01*

Three pieces fit together to make an agent:

- `Agent`

- `Runner.run()`

- `result.final_output`

Change the instructions and the same model can respond very differently.


---

## 🔧 Setup

```python
from pathlib import Path
from dotenv import load_dotenv
load_dotenv(dotenv_path=Path("..") / ".env")

from agents import Agent, Runner

MODEL = "gpt-5-mini"       # Course default, all standard agents
REASONING_MODEL = "gpt-5"  # Complex planning and reasoning tasks (see Lesson 29)

print("✅ Ready!")
```

---

## 🚀 How an Agent Works

Run the cell below to make your first deliberate agent call and see all three pieces in action.

```python
message = "What is Python?"

agent = Agent(
    name="Assistant",
    instructions="You are a helpful assistant.",
    model=MODEL
)

result = await Runner.run(agent, input=message)

print(result.final_output)
```

### 💡 Key Takeaway

`Agent` sets the name, instructions, and model.

`Runner.run()` executes the agent.

`result.final_output` holds the response.

---

## 🧪 Part 1: Instructions Shape Behavior

#### Run Concise Agent

```python
agent_concise = Agent(
    name="Concise Assistant",
    instructions="Answer every question in exactly one sentence.",
    model=MODEL
)

concise_result = await Runner.run(agent_concise, input=message)

print("📌 Concise:")
print(concise_result.final_output)
```

#### Run Friendly Agent

```python
friendly_instructions = (
    "Explain concepts to a 10-year-old using a fun analogy. "
    "Keep the answer to one short paragraph."
)

agent_friendly = Agent(
    name="Friendly Assistant",
    instructions=friendly_instructions,
    model=MODEL
)

friendly_result = await Runner.run(agent_friendly, input=message)

print("📌 Friendly:")
print(friendly_result.final_output)
```

### 💡 Key Takeaway

`instructions` shape the agent's persona and behavior.

## 🔭 Part 2: Tracing

Agent runs are recorded in traces by default.

A trace shows the input, instructions, and final output from that run.

## 🎯 Key Takeaways

**A basic agent call uses three pieces:**

- `Agent`: configures the name, instructions, and model

- `Runner.run()`: executes the agent and returns a result

- `result.final_output`: holds the final response
<br>
<br>

**Instructions shape behavior:**

- The same model and input can produce very different outputs when instructions change
<br>
<br>

**Tracing helps explain behavior:**

- The SDK records agent runs in traces by default
