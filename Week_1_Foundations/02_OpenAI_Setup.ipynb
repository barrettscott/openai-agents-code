## 📁 Step 4: Create Your .env File

Create a `.env` file in the course root so the notebook can load your API key.

All we need is a file named `.env` in the course root containing `OPENAI_API_KEY=sk-…`. Use whichever method feels easiest.

**Where it goes:**
```
📁 openai-agents/
├── 📁 Week_1_Foundations/
│   └── 📓 02_OpenAI_Setup.ipynb  ← You are here
├── 🔐 .env               ← We'll create this here
└── 📄 environment.yml
```
<br>

💡 **Note:** Files starting with `.` are hidden by default on macOS and Linux.
- **macOS:** In Finder, press **Cmd + Shift + .** to toggle hidden files

- **Linux:** In your file manager, press **Ctrl + H** to toggle hidden files

- **Windows:** Hidden files starting with `.` are visible by default



### ⚙️ Option A: Automated Setup

Run the cell below. When prompted, paste your API key and press Enter.

*Already have a `.env` file? The cell will detect it and ask if you want to keep or replace it.*

*Input box not appearing? Use Option B below.*

````python
from pathlib import Path
from getpass import getpass

# Path to .env file (in parent directory)
env_file = Path("..") / ".env"

print("Creating .env file...\n")
print(f"Location: {env_file.resolve()}\n")

# Check if .env already exists
if env_file.exists():
    print("ℹ️  .env file already exists!")
    print(f"   Location: {env_file.resolve()}\n")
    
    response = input("Do you want to replace it? (yes/no): ").strip().lower()
    
    if response not in ['yes', 'y']:
        print("\n✅ Keeping existing .env file.")
        print("   If you need to update it later, delete it manually and run this cell again.")
    else:
        env_file.unlink()  # Delete the file
        print("\n✅ Old .env file deleted. Creating new one...\n")

# If we get here, either file didn't exist OR user chose to replace it
if not env_file.exists():
    api_key = getpass("Paste your API key (hidden): ").strip()
    
    # Validation
    if not api_key:
        print("\n❌ No API key entered.")
        print("   Run this cell again and paste your key.")
    elif not api_key.startswith("sk-"):
        print("\n⚠️ Warning: API key doesn't start with 'sk-'")
        print("   Make sure you copied the entire key!")
    else:
        # Create .env file
        with open(env_file, 'w') as f:
            f.write(f"OPENAI_API_KEY={api_key}\n")
        
        print("\n✅ .env file created successfully!")
        print(f"   Saved to: {env_file.resolve()}")
        print("\n🔒 Your API key is stored securely.")
````

## 🔍 Step 5: Verify Your Setup

Let's verify that the notebook can load your API key from `.env`.

````python
import os
from pathlib import Path
from dotenv import load_dotenv

print("Verifying .env file...\n")

# Load .env file
env_path = Path("..") / ".env"
load_dotenv(dotenv_path=env_path, override=True)  # dotenv_path: location of .env; override=True: new key replaces any old one in memory

# Check if API key exists
api_key = os.getenv("OPENAI_API_KEY")

if not api_key:
    print("❌ API key not found!\n")
    print("Troubleshooting:")
    print("1. Make sure you ran the previous cell OR created .env via terminal")
    print("2. Check that .env file exists in the openai-agents folder")
    print("3. Check .env contains: OPENAI_API_KEY=your-key")
    print("4. No spaces around the = sign")
    print("5. Windows users: recreate .env using the PowerShell command in Step 4")
elif "your-actual-key-here" in api_key or api_key == "sk-proj-your-actual-key-here":
    print("❌ Placeholder key detected!\n")
    print("You need to replace 'sk-proj-your-actual-key-here' with your real API key.")
    print("Go back to Step 3 and paste your actual key from OpenAI.")
elif not api_key.startswith("sk-"):
    print("⚠️ API key doesn't start with 'sk-'\n")
    print("Make sure you copied the entire key from OpenAI")
else:
    # Show masked key
    masked_key = f"{api_key[:8]}...{api_key[-4:]}"
    print(f"✅ API key loaded: {masked_key}")
    print(f"✅ Key length: {len(api_key)} characters")
    print("\n🔒 Your full key is hidden for security")


    print("\n✅ Ready to test the connection!")
````

## 🚀 Step 6: Verify Your Connection with Your First Agent Call

Don't worry about understanding the code below yet — Lesson 03 walks through `Agent`, `Runner.run()`, and `result.final_output` piece by piece. For now, just run it and confirm you see "Setup complete!" You'll see this same shape — load key → create Agent → run — in every notebook from here on. Lesson 03 walks through it in detail.

````python
from pathlib import Path
from dotenv import load_dotenv
load_dotenv(dotenv_path=Path("..") / ".env", override=True)

from agents import Agent, Runner

MODEL = "gpt-5-mini"  # Course default — all standard agents

agent = Agent(
    name="SetupTest",
    instructions="You are a helpful assistant.",
    model=MODEL
)

try:

    result = await Runner.run(agent, input="Say 'Setup complete!' and nothing else.")

    print("🎉 SUCCESS! Your first agent response:\n")
    print(result.final_output)
    print("\n✅ You're ready for the course!")

except Exception as e:
    print(f"❌ Error: {e}\n")
    error_msg = str(e).lower()
    if "authentication" in error_msg or "api key" in error_msg:
        print("→ Check your API key is correct in .env")
    elif "billing" in error_msg or "quota" in error_msg:
        print("→ Add credits: https://platform.openai.com/account/billing/overview")
    elif "rate limit" in error_msg:
        print("→ Wait 60 seconds and try again")
    else:
        print("→ See troubleshooting section below")
````
