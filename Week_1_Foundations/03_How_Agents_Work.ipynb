## 🔧 Setup

```python
from pathlib import Path
from dotenv import load_dotenv
load_dotenv(dotenv_path=Path("..") / ".env")

from agents import Agent, Runner

MODEL = "gpt-5-mini"       # Course default — all standard agents
REASONING_MODEL = "gpt-5"  # Complex planning and reasoning tasks (full framework: Lesson 31)

print("✅ Ready!")
```

---

## 🚀 How an Agent Works

Run the cell below to make your first deliberate agent call and see all three pieces in action.

```python
message = "What is Python?"

agent = Agent(
    name="Assistant",
    instructions="You are a helpful assistant.",
    model=MODEL
)

result = await Runner.run(agent, input=message)

print(result.final_output)
```

### 💡 Why This Works

You just used the three core pieces: `Agent` defines who and how, `Runner.run()` executes it, and `result.final_output` is the response. Every notebook in this course follows this pattern.

---

## 🧪 Part 1: Instructions Shape Behavior

#### Run Concise Agent

```python
agent_concise = Agent(
    name="Concise Assistant",
    instructions="Answer every question in exactly one sentence.",
    model=MODEL
)

concise_result = await Runner.run(agent_concise, input=message)

print("📌 Concise:")
print(concise_result.final_output)
```

#### Run Friendly Agent

```python
friendly_instructions = (
    "Explain concepts to a 10-year-old using a fun analogy. "
    "Keep the answer to one short paragraph."
)

agent_friendly = Agent(
    name="Friendly Assistant",
    instructions=friendly_instructions,
    model=MODEL
)

friendly_result = await Runner.run(agent_friendly, input=message)

print("📌 Friendly:")
print(friendly_result.final_output)
```

### 💡 Why This Works

The only change was `instructions`, which defines the agent's persona and behavior for every run. This is the most powerful lever you have — good instructions decide how your agent behaves when a real task is ambiguous, risky, or needs a tool.

## 🔭 Part 2: Tracing

Every agent run is automatically traced — no configuration needed. You can inspect a trace in the OpenAI platform to see the input, the agent instructions, and the final output together. When an agent does something unexpected, this is the first place to look — traces show you exactly what input and instructions produced that output, so you don't have to guess.

After running the demos above, open the Traces page in your OpenAI dashboard to see the runs you just created: [platform.openai.com/traces](https://platform.openai.com/traces)
