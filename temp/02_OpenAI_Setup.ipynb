## 📁 Step 2: Create Your .env File

### What is a .env file?

A `.env` file stores your API key securely:
- Keeps secrets out of your code
- Standard practice in professional development
- Never gets committed to version control

**Where it goes:**
```
📁 claude-agents/
├── 📁 Week_1_Foundations/
│   └── 📓 02_Anthropic_Setup.ipynb  ← You are here
├── 🔐 .env               ← We'll create this here
└── 📄 environment.yml
```

💡 **Note:** Files starting with `.` are hidden by default on macOS and Linux.
- **macOS:** In Finder, press **Cmd + Shift + .** to toggle hidden files
- **Linux:** In your file manager, press **Ctrl + H** to toggle hidden files
- **Windows:** Hidden files starting with `.` are visible by default

---

````python
from pathlib import Path
from getpass import getpass

# Path to .env file (in parent directory)
env_file = Path("..") / ".env"

print("Creating .env file...\n")
print(f"Location: {env_file.resolve()}\n")

# Check if .env already exists
if env_file.exists():
    print("ℹ️  .env file already exists!")
    print(f"    Location: {env_file.resolve()}\n")
    
    response = input("Do you want to replace it? (yes/no): ").strip().lower()
    
    if response not in ['yes', 'y']:
        print("\n✅ Keeping existing .env file.")
        print("    If you need to update it later, delete it manually and run this cell again.")
    else:
        env_file.unlink()  # Delete the file
        print("\n✅ Old .env file deleted. Creating new one...\n")

# If we get here, either file didn't exist OR user chose to replace it
if not env_file.exists():
    api_key = getpass("Paste your API key (hidden): ").strip()
    
    # Validation
    if not api_key:
        print("\n❌ No API key entered.")
        print("    Run this cell again and paste your key.")
    elif not api_key.startswith("sk-ant-"):
        print("\n⚠️ Warning: API key doesn't start with 'sk-ant-'")
        print("   Make sure you copied the entire Anthropic key!")
    else:
        # Create .env file
        with open(env_file, 'w') as f:
            f.write(f"ANTHROPIC_API_KEY={api_key}\n")
        
        print("\n✅ .env file created successfully!")
        print(f"    Saved to: {env_file.resolve()}")
        print("\n🔒 Your API key is stored securely.")
````

## 🔍 Step 3: Verify Your Setup

Let's check that your .env file was created correctly:

````python
import os
from pathlib import Path
from dotenv import load_dotenv

print("Verifying .env file...\n")

# Load .env file
env_path = Path("..") / ".env"
load_dotenv(dotenv_path=env_path, override=True)

# Check if API key exists
api_key = os.getenv("ANTHROPIC_API_KEY")

if not api_key:
    print("❌ API key not found!\n")
    print("Troubleshooting:")
    print("1. Make sure you ran the previous cell OR created .env via terminal")
    print("2. Check that .env file exists in the claude-agents folder")
    print("3. Check .env contains: ANTHROPIC_API_KEY=your-key")
    print("4. No spaces around the = sign")
    print("5. Windows users: recreate .env using the PowerShell command in Step 2")
elif "your-actual-key-here" in api_key or api_key == "sk-ant-your-actual-key-here":
    print("❌ Placeholder key detected!\n")
    print("You need to replace 'sk-ant-your-actual-key-here' with your real API key.")
    print("Go back to Step 2 and paste your actual key from the Anthropic Console.")
elif not api_key.startswith("sk-ant-"):
    print("⚠️ API key doesn't start with 'sk-ant-'\n")
    print("Make sure you copied the entire key from the Anthropic Console")
else:
    # Show masked key
    masked_key = f"{api_key[:8]}...{api_key[-4:]}"
    print(f"✅ API key loaded: {masked_key}")
    print(f"✅ Key length: {len(api_key)} characters")
    print("\n🔒 Your full key is hidden for security")


    print("\n✅ Ready to test the connection!")
````

## 🚀 Step 4: Verify Your Connection with Your First Agent Call

Let's test everything by asking the AI a question:

````python
from pathlib import Path
from dotenv import load_dotenv
load_dotenv(dotenv_path=Path("..") / ".env", override=True)

from claude_agent_sdk import query, ClaudeAgentOptions, AssistantMessage

MODEL = "claude-sonnet-4-5"

options = ClaudeAgentOptions(
    system_prompt="You are a helpful assistant.",
    model=MODEL
)

try:

    response_text = ""
    async for message in query(prompt="Say 'Setup complete!' and nothing else.", options=options):
        if isinstance(message, AssistantMessage):
            for block in message.content:
                if hasattr(block, "text"):
                    response_text += block.text

    print("🎉 SUCCESS! Your first agent response:\n")
    print(response_text)
    print("\n✅ You're ready for the course!")

except Exception as e:
    print(f"❌ Error: {e}\n")
    error_msg = str(e).lower()
    if "authentication" in error_msg or "api key" in error_msg:
        print("→ Check your API key is correct in .env")
    elif "billing" in error_msg or "credit" in error_msg:
        print("→ Add credits: https://console.anthropic.com/settings/billing")
    elif "rate limit" in error_msg:
        print("→ Wait 60 seconds and try again")
    else:
        print("→ See troubleshooting section below")
````

## 🔧 Troubleshooting

**Authentication error or API key not found?**
- Check you copied the entire key (starts with `sk-ant-`)
- Format: `ANTHROPIC_API_KEY=sk-ant-your-key` (no spaces around `=`)
- Check `.env` is in the `claude-agents` folder, not inside `Week_1_Foundations`

**Key not loading after editing `.env`?**
- Restart the kernel after editing `.env` — changes don't take effect until the kernel restarts
- Windows users: recreate `.env` using the PowerShell command in the Manual Alternative section

**Billing error (no credits)?**
- Add payment method at console.anthropic.com/settings/billing
- Minimum $5 credit required

**.env file not found?**
- File must be named exactly `.env` (starts with dot)
- Must be in the `claude-agents` folder, not inside `Week_1_Foundations`

**Connection or rate limit error?**
- Check internet connection and disable VPN temporarily
- Wait 60 seconds if you hit rate limits
- Check status.anthropic.com for outages

**Still having issues?**
- Copy any error message and paste it into Claude, ChatGPT, Gemini, or Grok — they're great at debugging setup issues
- Re-watch the setup lecture
- Post to the Q&A with your error message and output
