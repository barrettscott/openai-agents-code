## ✅ Check 1: Python Version

We need Python 3.10 or higher for this course.

Run the cell below and look for the green ✅ checkmark!

````python
import sys

print("Checking Python version...\n")

version_info = sys.version_info

if version_info.major == 3 and version_info.minor >= 10:
    print(f"✅ Python {version_info.major}.{version_info.minor} detected - this should work!")
else:
    print(f"❌ Python {version_info.major}.{version_info.minor} detected")
    print("    This course requires Python 3.10 or higher.")
    print("    Did you activate the conda environment?")
    print("    Run: conda activate claude-agents")
````

---

## 📦 Check 2: Required Packages

Let's verify all course packages are installed.

````python
print("Checking required packages...\n")

all_ok = True

try:
    import claude_agent_sdk
    print(f"✅ claude-agent-sdk {claude_agent_sdk.__version__} installed")
except ImportError:
    print("❌ claude-agent-sdk missing — run: conda env update -f environment.yml")
    all_ok = False

try:
    from dotenv import load_dotenv
    print("✅ python-dotenv installed")
except ImportError:
    print("❌ python-dotenv missing — run: conda env update -f environment.yml")
    all_ok = False

try:
    from pydantic import BaseModel
    print("✅ pydantic installed")
except ImportError:
    print("❌ pydantic missing — run: conda env update -f environment.yml")
    all_ok = False

if not all_ok:
    print("\n    To fix all missing packages:")
    print("    1. conda activate claude-agents")
    print("    2. conda env update -f environment.yml")
    print("    3. Restart JupyterLab")
````

---

## 🔧 Check 3: Jupyter Kernel

Verify you're using the correct conda environment in Jupyter.

````python
import sys

print("Checking Jupyter kernel...\n")

if "claude-agents" in sys.executable.lower() or "claude-agents" in sys.prefix.lower():
    print("✅ Correct conda environment active!")
else:
    print("⚠️ Warning: May not be using the claude-agents environment")
    print("\n    To fix:")
    print("    1. In JupyterLab menu: Kernel → Change Kernel")
    print("    2. Select 'Python 3 (ipykernel)' — the claude-agents environment")
    print("    3. Re-run this notebook")
````

## 🔧 Troubleshooting

**Python version wrong?**
- Make sure you ran: `conda activate claude-agents`
- Close and restart JupyterLab
- Check: Kernel → Change Kernel → Python 3 (ipykernel) (claude-agents)

**Packages missing?**
- Run: `conda activate claude-agents`
- Run: `conda env update -f environment.yml`
- Restart JupyterLab and restart the kernel

**Wrong kernel?**
- Kernel → Change Kernel → Python 3 (ipykernel) (claude-agents)
- Make sure it says "(claude-agents)" in your terminal

**Still having issues?**
- Copy any error message and paste it into Claude, ChatGPT, Gemini, or Grok — they're great at debugging setup issues
- Re-watch the setup lecture
- Post to the Q&A with your error message and output
