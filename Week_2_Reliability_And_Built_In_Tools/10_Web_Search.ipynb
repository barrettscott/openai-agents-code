## 🌐 Part 1: Enabling Web Search

Adding web search to an agent takes one line — `WebSearchTool()` in the `tools` list.

No extra search API key, no configuration required.

#### Without Web Search

```python
agent_no_search = Agent(
    name="NoSearchAgent",
    instructions="You are a helpful assistant.",
    model=MODEL
)

result = await Runner.run(agent_no_search, input="What are the top AI news stories this week?")

print("Without web search:")
print(result.final_output)
```

#### With Web Search

```python
agent_search = Agent(
    name="WebSearchAgent",
    instructions="You are a helpful research assistant. Always search the web for current information.",
    model=MODEL,
    tools=[WebSearchTool()]
)

result = await Runner.run(agent_search, input="What are the top AI news stories this week?")

print("With web search:")
print(result.final_output)
```

### 💡 Why This Works

`WebSearchTool()` is a built-in tool — OpenAI runs the search, no API key needed.

This **grounds** the agent's answer in real retrieved content instead of training data.

The result: better accuracy on time-sensitive questions and less hallucination.

⚠️ **Security note:** Web search results are untrusted input that can contain prompt-injection attacks (see Lesson 23: Prompt Injection & Tool Safety).

Treat all retrieved web content as potentially adversarial.

## 📎 Part 2: Citations

Web search responses can include inline citations — the agent includes them more reliably when you ask for them.

```python
agent_citations = Agent(
    name="CitationAgent",
    instructions="You are a research assistant. Always cite your sources clearly.",
    model=MODEL,
    tools=[WebSearchTool()]
)

result = await Runner.run(agent_citations, input="What is the current version of Python and when was it released?")

print("📎 RESPONSE — looking for inline citations")
print("=" * 60)
print(result.final_output)
print("=" * 60)
```

### 💡 Why This Works

Citations may appear automatically, but they aren't guaranteed.

If your application requires citations, instruct the agent explicitly to cite its sources.

Look for inline URL references in the text — those are the citations the agent surfaced from search results.

---

## 📍 Part 3: Location-Targeted Search

`WebSearchTool` accepts an optional `user_location` parameter to bias results toward a specific geography — useful for local news, weather, business searches, and anything region-specific.

```python
# Change these to your location
# "approximate" is the only supported type for user_location
CITY = "Denver"
REGION = "Colorado"
COUNTRY = "US"

agent_local = Agent(
    name="LocalSearchAgent",
    instructions="You are a helpful local information assistant.",
    model=MODEL,
    tools=[WebSearchTool(
        user_location={
            "type": "approximate",
            "city": CITY,
            "region": REGION,
            "country": COUNTRY
        }
    )]
)

result = await Runner.run(agent_local, input="What is the weather like today?")

print("📍 LOCAL SEARCH RESULT")
print("=" * 60)
print(result.final_output)
print("=" * 60)
```

### 💡 Why This Works

Without a location hint the model guesses at context.

With `user_location`, results are biased toward that geography — broad queries may still return global results.
